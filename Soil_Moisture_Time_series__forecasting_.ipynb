#Abstract


Soil moisture refers to the water stowed in the soil and is influenced by factors such as precipitation, temperature, and soil elements. The type of biome, as well as the suitableness of land for producing crops, are determined by the same criteria. 
The crops' health is dependent on a sufficient supply of soil nutrients and moisture, among other elements. Plants' regular function and growth are interrupted as moisture availability decreases, resulting in lower crop yields.

Soil moisture prediction can be extremely useful in a variety of applications, such as precision agriculture and ecosystem management. # Introduction

Soil moisture is defined differently in various disciplines. A farmer, for example, has a different understanding of soil moisture than a  climate forecaster or water resource management. Soil moisture is typically defined as the water contained in the gaps between soil particles. 
The water available in the top 10 cm of soil is called surface soil moisture, and the water in the top 200 cm of soil, which is generally assumed to be the level that is available to the plants, is called root zone soil moisture.


Although soil moisture has a minor volume in comparison to other components of the hydrologic cycle, it is essential to numerous biological, hydrological, and biogeochemical processes. Weather,  water quality, soil erosion, runoff possibility, flood control, reservoir control, and geotechnical engineering are all areas where soil moisture information is useful to both government and private organizations.


As we mentioned, soil moisture has a great impact on the Climate and weather industries, by playing a significant role in the development of weather patterns, and precipitation production.
This is because Soil moisture is an important factor in determining how much water and heat energy is exchanged between the land level and the atmosphere via vaporization and plant transpiration. 
Improved characterization of surface soil moisture, temperature, and vegetation can lead to considerable forecast advancements, according to simulations using numerical weather prediction models.(1)

In agricultural industry, the allocation of water between different segments of the hydrological cycle is heavily influenced by soil moisture. It aids in the comprehension of hydrology and climatic circumstances with significant geographical and temporal variability. It also impacts essential farm processes such as selection of the crop, tilling timing, planting, fertilizer usage, harvesting, vaporization, runoff, and heat.(2,3)
Soil moisture measurement and forecasting delivers information about desired infiltration, and runoff production throughout rainfall events, as well as water management for agricultural uses.(4)

# Artificial Antelligence




To develop models and produce predictions utilizing algorithms and other methods, data scientists depend extensively on artificial intelligence(AI), particularly its subfields which are known as machine learning and deep learning.
## Traditional Machine learning algorithm
The machine learning algorithm is a subfield of artificial intelligence (AI). The term "machine learning" refers to when algorithms learn from data. In machine learning, algorithms use previous data to learn, predict outcomes as new data enters, and identify patterns that humans would miss, without being explicitly trained. rather, they identify and learn the patterns in the provided data and generate predictions based on what they learned as new information comes in.

Simply said, the machine learning algorithm learns by examples and enables users to utilize those self-learning algorithms to reveal insights, specify relationships, and forecast forthcoming trends. Healthcare, energy, marketing, financial technology, and many more industries use the functional implications of machine learning. Machine learning, when used correctly, ables businesses to find the best answers to actual challenges, resulting in real, measurable commercial value.

A linear regression model could be the simplest model for traditional machine learning.

Depending on the data utilized to feed the algorithms, the learning process for machine learning algorithms can be divided into two procedures supervised learning and unsupervised learning.

**Supervised learning**, uses the labeled data to supervise or train the algorithms so that they can accurately classify data or forecast outcomes. In this approach, the model tracks its accuracy and learns over time by using labeled inputs and outputs.

When it comes to data mining, supervised learning may be divided into two categories: **classification** and **regression**.

***Classification*** approaches employ an algorithm to precisely assign test data into certain classes, such as segregating zero from one.

Classification algorithms can be implemented in different types such as Logistic Regression, K-nearest Neighbors, decision trees, random forest, Naive Bayes, and support vector machines(SVM).


***Regression*** problems, employ an algorithm to comprehend the relationship betwixt dependent and independent featurs. Regression models can be used to predict numerical values relying on several data points. linear regression, polynomial regression and logistic regression are some of the common regression algorithms.


**Unsupervised learning** analyzes and clusters unlabeled data utilizing machine learning algorithms. The term "unsupervised" basically comes from the fact that these algorithms find concealed patterns in data without human interposition.

Unsupervised approaches are mostly used for **Clustering**, **association**, and **dimension reduction**.


***Clustering*** is a data mining method that groups unlabeled data into groups based on resemblances and distinctions.



***The Association*** method employs several rules to discover associations betwixt variables in a dataset.


***The dimensionality reduction*** method is employed when the number of dimensions(features) in a dataset is too large. The dimensionality reduction method keeps the data integrity while reducing the dimension of data inputs to a reasonable size.

## The Deep learning algorithm

Deep learning is a subfield of machine learning and it refers to algorithms that evaluate data using a logic structure equivalent to a human. This can occur through both supervised and unsupervised learning.

Deep learning applications accomplish this by employing an artificial neural network (ANN) Also called neural network (NN) which is a layered structure of algorithms. The Input is passed through numerous "layers" of neural network algorithms, each layer passes a clarified version of the data to the next layer, in order to process and compute big data.

The majority of machine learning algorithms perform effectively on datasets of a few hundred features or columns.
As the data input passes via each neural network layer, deep learning algorithms learn more about it. Primary layers learn to identify low-level elements, while later layers combine those features into a more holistic outcome.

The structure of an artificial neural network is based on the biological neural network of the brain of a human, resulting in a significantly more capable learning process than traditional machine learning models.

Deep learning's capacity to process vast amounts of features makes it incredibly strong, especially when processing unstructured data.  On the other hand, the algorithms may be overused for less complicated tasks because they require massive amounts of data to be practical.






Neural networks can be devided into several types,  each of which serves a different purpose.
The perceptron with a single neuron is a simplest form of neuron network.
[Reference](https://www.ibm.com/cloud/learn/neural-networks)

[Reference](https://www.ibm.com/cloud/learn/recurrent-neural-networks)

**Multi-Layer Perceptrons MLPs** also known as feedforward networks are classical type of neuron network and   include of different layers: "an input layer, a hidden layer or layers, and an output layer". althogh these neural networks are known as MLPs, it's vital to remember that they're made up of sigmoid neurons rather than perceptrons because most real-world problems are not linear. MLPs are the base for computer vision, natural language processing and other artificial neural networks.

**Convolutional Neural Networks (CNNs)** are same as MLPs Neural networks, however CNNs used mostly for pattern identification, image recognition, and/or computer vision. These networks use linear algebra principles, notably matrix multiplication, to find patterns in images.


**Recurrent Neural Networks (RNNs)** are a form of artificial neural network that are determined by their feedbackloops and commonly used for making predictions in time-series data, such as stock market forecasting or sales forecasting. In this study we are using RNNs neural network.

**Bidirectional recurrent neural networks (BRNN), Gated recurrent units (GRUs), and Long short-term memory (LSTM)** are Variant architectures of RNN, in This study we are using **LSTM** to forecast the soil moisture value 30 days ahead.
**LSTM** is a well-known RNN architecture that was developed by Sepp Hochreiter and Juergen Schmidhuber to solve the vanishing gradient problem,which could be encountered when training traditional RNNs.

**LSTM** architecture are the most suitable networks to classify, process, and make predictions based on time series data, as there might be lags of uncertain duration between critical occurrences in a time series.


In the hidden layers of the neural network, LSTMs have "cells" that include three gates: an input gate, an output gate, and a forget gate. These gates regulate the flow of data required to forecast the network's output.

# Time Series Components


Dividing a time series into systematic and unsystematic elements is a helpful abstraction for finding the most appropriate forecasting methods.

**Systematic:** Elements of a time series that have constancy or repetition which could be represented and modeled are known as systematic.

**Non-Systematic:** Time series elements that can't be straightly modeled are called non_systematic.

A time series is assumed to include three systematic components: seasonality, trend, and level, as well as one non-systematic component known as noise.

[References](https://machinelearningmastery.com/decompose-time-series-data-trend-seasonality/#:~:text=Trend%3A%20The%20increasing%20or%20decreasing,term%20cycle%20in%20the%20series.)


## Trend
A trend in data is a pattern that demonstrates the movement of a series to progressively greater or lesser values over time. That is to say when the time series has an increasing or decreasing slope it's called a trend. A trend usually lasts for a while before disappearing; it does not repeat itself.

Trends could be observed as:

**Uptrend:**  When a time Series Analysis displays an upward general tendency, it is called an uptrend.

**Downtrend:**  When Time Series Analysis displays a downward tendency, it is called a downtrend.


**Horizontal or Stationary trend:** This type of trend occurs when there is no discernible pattern.

[References](https://www.geeksforgeeks.org/what-is-a-trend-in-time-series/)

## Seasonality
Seasonality in time series data points out to the existence of fluctuations that occur at regular intervals, such as weekly, monthly, or even quarterly however never a year. Seasonality can be caused by a variety of circumstances, including, weather, holidays, and vacations. They consist of predictable patterns that are periodic, repeated, and generally regular.

[References](https://thebusinessprofessor.com/en_US/mgmt-operations/seasonality-forecast-definition)

In addition to seasonal influences, another predictable fluctuations with no definite period can be found in time series. A cyclic pattern describes this type of change.

## Residual

After the seasonal and trend components of a time series have been assessed and removed, the residual component commenly known as the irregular componant is what is left. Short-term fluctuations in the series cause it, and they are neither systematic nor predictable. These fluctuations can overwhelm movements in a highly irregular series and masking the trend and seasonality.


Residuals are crucial when defining the quality of a model. We can look at residuals to see how big they are and if they create a pattern.
The model predicts precisely when all of the residuals are zero. The model becomes less accurate as the residuals get further away from zero. In linear regression,  the higher the aggregate of squared residuals, the less the R-squared statistic and the rest being equal.

When the average residual is greater or less than 0, it indicates that the model is systematically biased, which means, it is always over-or under-predicting.

When residuals include patterns, the model is qualitatively incorrect because it fails to explain several properties of the data. Most statistical tests are invalidated by the presence of patterns.


## Combining the Time series component

A series is assumed to be a total or mixture of the mentioned four components.
There is a level and noise in every series. The seasonality and trend elements are arbitrary. It's beneficial to consider the components combined as additively or multiplicatively.

### Additive Model

According to an additive model, the components are added to each other as shown below:

```y(t) = Level + Trend + Seasonality + Noise```

This model is linear, one in which modifications are consistently produced by the same value across time.

A straight line is a linear trend.
The frequency and amplitude of linear seasonality are the same.


### Multiplicative Model

According to a multiplicative model, the components of the time series are multiplied by each other as shown below:

``` y(t) = Level * Trend * Seasonality * Noise ```



This model is nonlinear. like a quadratic, or exponential. Over time, changes rise or diminish.

A curved line represents a nonlinear trend.

The frequency and/or amplitude of a non-linear seasonality increases or decreases over time.

# Stationary 




In time series, stationary is a crucial factor. For instance, the first step in ARIMA time series forecasting model is to identify the number of differencing needed to construct the stationary series, as the ARIMA model cannot forecast on non-stationary time series.

The statistical attributes, such as variance, mean and covariance, do not change with time or are independent of the time at which it is observed. Therefore, time series with seasonality or trends are  considered a not stationary; the trend and seasonality will impact the time series' value at different times that simply means that the seasonality and trend effect the value of the time series at variant times. 

Stationarity is significant as it is required by numerous valuable analytical tools, statistical tests, and models.

As a result, knowing whether or not a time series is stationary is crucial. Rather than choosing between two strict possibilities, this usually entails being capable to determine that a series is created by a stationary process with a high degree of certainty.


The majority of forecasting approaches hypothesize that the distribution has stationary. Autocovariance and autocorrelations, for instance, are based on the premise of stationarity. Unexpected or odd behaviors, such as t-ratios that don't follow a t-distribution or high r-squared values ascribed to features that aren't associated at all, can be caused by a lack of stationarity.


## Types of stationarity 
 
**Strict Stationarity:** The common distribution of any times of any degree for example anticipated variances, values, third-order, and greater times, through the process is never time-dependent. In practice, this type of stationery is far too stringent to be applied to any real-world model.

**First-order Stationary:** In the first-order stationarity series, the means do not ever vary over time. however Other statistics, such as variance, are subject to change.

**Second-order stationarity:** Also known as weak stationarity has a steady variance, mean, and autocovariance over time. However other statistics can modify in time. Second-order stationarity is a limited issue of strict stationarity and it is fairly common.

**Terend Stationary:** Models that are trend-stationary, fluctuate near the mean of the series which is known as a deterministic trend. These deterministic trends may be quadratic or linear. However, the amplitude of the oscillations or the height of one fluctuation does not change over time series.

**Difference Stationary:** Models that require one or further differencing to be considered as stationary are known as difference-stationary models.


The soil moister series is a an example of difference stationary, we will see how it will transform to a stationatry time series with one differnecing later in this study.
## Differencing

Differencing is a time series dataset transformation method.

It can be used to eliminate time dependency of the series, known as temporal dependence. Trends and seasonality are examples of such structures.

Differencing removes fluctuations in the level of timeseries and decrease seasonality and trend, which will result in reinstating the mean of the time series.

In a simple word the difference method is calculating the differences betwixt sequential observations, which will result the traformation of a non-stationary time series to stationary.
In a simple word the difference method is calculating the differences betwixt sequential observations,and, which will result the traformation of a non-stationary time series to stationary.


$$
Difference(t) = Observation(t) - Observation(t-1)
$$



In some cases like a nonlinear trend, the temporal structure might yet remain after implementing the differencing method. In such cases, we can repeat the differencing process until all time dependencies are removed.
The **difference order** refers to the number of times differencing is conducted.

A **lag-1 difference** is a difference betwixt two sequential observations.
The number of lag differences can be customized to fit the temporal structure. in the case of seasonality, it would probably be the seasonality width(period).



In this study we are using the diff() method from pandas library to implement the differencing on soil moister time series. this method includes parameters such as period(cycles to shift in order to compute the difference), axis (to specify differencing over columns(1) or rows(0)(default value is 0)).

# Method to check the stationarity of the Time Series Data


## Augmented Dickey- fuller Test (ADCF):



An augmented Dickey-Fuller test uses the null and alternative hypotheses to determine whether a time series is stationary.
these null and alternative hypotheses are :

**H0:** The time series has some time-dependent architecture and does not include constant variance in time. In simple words the time series is non-stationary.

**H1:** The time series is stationary.


We can reject the null hypothesis and infer that the time series is stationary if the p-value from the Dickey-Fuller test is smaller than a certain significance level (by default α =0.05).


[References](https://www.statology.org/dickey-fuller-test-python/)

## KPSS TEST

The KPSS test is a statistical method and a type of Unit root test,for determining if a time series is stationary near a deterministic trend. The KPSS test, same as the ADF test, is often used to determine whether a series is stationary. but In terms of function and practical application, it differs significantly from the ADF exam.

The test is relatively similar to the ADF test,  however, it is a frequent fallacy that can be used interchangeably with the ADF test. This can cause stationarity misinterpretations, which can simply go unsought and cause more troubles down the road.

The statsmodel module in Python provides a simple implementation of the KPSS test.
The null hypothesis of the KPSS test is that the series is stationary, which is the key contrast to the ADF test.
So, basically, p-value interpretation is the opposite of each other in KPSS and ADF tests.

The series is non-stationary if the p-value is below a certain threshold (for example, 0.05). In the ADF test, however, it indicates that the time series being examined is stationary.
### Differneces of KPSS test & ADF test

The capacity of the KPSS test to check for stationarity in the 'existence of a deterministic trend' is a significant difference between it and the ADF test.

As we mentioned above, the KPSS checks for series stationarity around a 'deterministic trend.'

That means that even if a series is gradually increasing or decreasing, the test may not necessarily reject the null hypothesis (that the series is stationary).
#### what is a ‘deterministic trend’?
The term deterministic suggests that the trend in the series does not alter over time. in other words, even if the series is shocked, it usually returns to its former direction.

### Implementation of the KPSS Test


We'll utilize the statsmodel's kpss method to implement the KPSS test. The code below runs the test and prints the results as well as the result's interpretation. There 4 values in the printed result that need to be considered for KPSS test. The **KPSS statistic**, **p-value**, **Number of lags that the test utilize** and **Critical values**

**The p-value** is a probability score that helps us to decide whether or not to reject the null hypothesis. We reject the null hypothesis if the p-value is less than a predetermined threshold, which is by default 0.05.


**The KPSS statistic** is the test statistic that is calculated as the test is being performed. The test statistic value must be bigger than the critical values to reject the null hypothesis. If it is indeed bigger than the target critical value, so the p-value is automatically low.
In other words, The kpss statistic will be bigger than the 5% critical value when the p-value is smaller than the null hypotheses(0.05).

**The number of lags** is the number of lags in the series that the kpss test model actually utilized. The 'legacy' technique is used by default in statsmodels kpss(). In this method,  where n is the length of the series, the number of lags is:
$$int(12 * (n / 100)**(1 / 4) $$


[References](https://www.machinelearningplus.com/time-series/kpss-test-for-stationarity/)

````python
df_by_day['BF10'].plot()
````

````python
from statsmodels.tsa.stattools import kpss
def kpss_test(series, **kw):    
    statistic, p_value, n_lags, critical_values = kpss(series, **kw)
    # Format Output
    print(f'KPSS Statistic: {statistic}')
    print(f'p-value: {p_value}')
    print(f'num lags: {n_lags}')
    print('Critial Values:')
    for key, value in critical_values.items():
        print(f'   {key} : {value}')
    print(f'Result: The series is {"not " if p_value < 0.05 else ""}stationary')

kpss_test(df_by_day["BF10"])
````

As we can see in the printed out put above, the p-value is 0.01, which is less than the null hypothesis 0.05, hence the series is **non-stationary**. 

But considering the the'stationarity around a deterministic trend,' we believed that the KPSS should not reject the null. well, the reason for that is beacouse the KPSS by deafult test for sationary around the mean, and in order to test the series around the trend we need to explicitly pass the parameter ```regression='ct' ``` to KPSS, as below:



````python
kpss_test(df_by_day['BF10'], regression='ct')
````

# Autocorrelation



Autocorrelation is the degree of resemblance between The particular time series which we studying and a lagged history of the variable itself over subsequent time gaps. In other term, autocorrelation is used to determine the link between the current value of a variable and any previous values to which you have access.


The term auto means self, and autocorrelation particularly analyse and compare a variable's present values to its historical data. It plots one series against the other and assesses how similar they are. A plot of autocorrelation is used to determine whether the components of a time series are positively, negatively, or independently correlated. 

For the purpose of comparison, autocorrelation is effectively the same process as computing the correlation between two separate sets of time series values. The main distinction here is that autocorrelation utilizes the same time series twice: one time in the variable's original values and another time after a few various time periods have happened.

Serial correlation, lagged correlation and time-series correlation are all terms used to describe autocorrelation. An autocorrelation is an excellent approach for revealing trends and patterns in time series data that would otherwise go unnoticed.

## Importance of Autocorrelation 

Regression analysis is frequently one of the initial steps in any data study. However, having no autocorrelation in data is one of the hypotheses of regression analysis. This can be aggravating since attempting to perform a regression analysis on data with autocorrelation can result in inaccurate results.

Furthermore, several time series forecasting approaches (particularly regression modeling) assume that the residuals(the difference between the fitted model and the data)are free of autocorrelation . People frequently utilize residuals to determine whether their model is a well fit, neglecting the hypothesis that the residuals have no autocorrelation. This oversight may lead people to believe that their model is a suitable fit while it isn't.

Finally, possibly the most convincing part of autocorrelation analysis is how it can aid in the discovery of hidden patterns in data and the selection of appropriate forecasting methods. We may use it to discover seasonality and trend in the time series data in particular. Additionally, selecting the optimal ARIMA model for the time series prediction requires examining the autocorrelation function (ACF) and partial autocorrelation function (PACF) together.

##Partial Autocorrelation Function (PACF)


A partial correlation is a conditional correlation. It is the correlation among two featurs with a hypothesis that values of other set of featurs are known and and considered.

Consider the following regression scenario: x1, x2, and x3 are the predictor featurs, and y is the response(target) feature. 
The partial correlation betwixt x3 and y is calculated by considering how both y and x3 are correlated to x2 and x1. This partial correlation can be discovered  by correlating the residuals of two separate regressions:

1. In the first regression We take x1 and x2 into account to forecast y.
2. in Seconde regression we take x1 and x2  into account to predict x3. We simply correlate the "parts" of y and x3 that x2 and x1 didn't predict.

### Partial Autocorrelation in Time Series




In time_seris, the partial autocorrelation(PACF) is the size of the correlation betwixt samples of a time_series which are independent of each other by k time(Yt and Yt-k), after modifying for the existence of every other terms of smaller lag (yt–1, yt–2,..., yt–k–1).

### Interpretation with PACF:
ARIMA models can be identified by the partial autocorrelation and autocorrelation functions. The partial autocorrelation function should show the following patterns. Assess the spikes of every lag in order to specify they are significant or not. The spike which extends above the confidence interval is considered a significant spike, indicating that the lag correlation is not zero.

1. At lag 1, there is a large spike that falls after a few lags:

This pattern in PCAF indicates a movement of the average term in the data. To establish the order of the moving average term, utileze the autocorrelation method.

2. Great spike at lag 1, go along with a damped wave with positive and negative correlations that alternates:


This pattern indicates a wider moving average term in the dataset. To establish the order of the moving average term, use the autocorrelation function.


3. Significant correlations appear at the first or second lag, and then non-significant correlations occur after that:

This pattern indicates that In the data, there is an autoregressive term. The order of the autoregressive term is determined by the sum of significant correlations.

### Improving AR Model 

Another option that I am using here to improve our AR model is to manually make predictions using the learned coefficients. This necessitates keeping the history of 1 (number of lags)  previous observation and retrieving the coefficients from the model to utilize in the regression equation to generate fresh forecasts.


[Reference](https://machinelearningmastery.com/autoregression-models-time-series-forecasting-python/)

````python
X = df_by_day['BF10'].values
train= X[:len(X)-30]
test = X[len(X)-30:]
window = 1
model = AutoReg(train, lags=1).fit()
coef = model.params
# walk forward over time steps in test
history = train[len(train)-window:]
history = [history[i] for i in range(len(history))]
pred = list()
for t in range(len(test)):
	length = len(history)
	lag = [history[i] for i in range(length-window,length)]
	yhat = coef[0]
	for d in range(window):
		yhat += coef[d+1] * lag[window-d-1]
	obs = test[t]
	pred.append(yhat)
	history.append(obs)
	print('predicted=%f, expected=%f' % (yhat, obs))
rmse = sqrt(mean_squared_error(test, pred))
print('Test RMSE: %.3f' % rmse)

print(model.summary())
pyplot.plot(test)
pyplot.plot(pred, color='red')
pyplot.show()
````

As we can see the model has improved significantly and it is predicting qyute accurate. The RMSE has also reduced to  5.079.

## A seasonal autoregressive integrated moving average with eXogenous factors(SARIMAX)




The issue with the ARIMA model is that it does not account for seasonality.
In the case of time series with seasonality, the extended form of ARIMA model, which is SARIMA model is a better choice as it include seasonal differencing.

Seasonal differencing is same as differencing, but in season differencing we remove the value from the prior season instead of consecutive terms substraction.
The (SARIMA) model differs from an ARIMA model based on seasonal patterns by one step. 

The SARIMA model, can correct a non-stationary time series by removing trend and seasonality.
Since th BF10 seies also has seasonility so this model could be a better choice than ARIMA to forecast.
SARIMA model includes 7 parameters as:

**order=(p,d,q)** which we already dicussed how to define them in ARIMA model.
**Seasonal_order=(P,D,Q)s** where **P** is the order of AR seasonilty, **D** is the number of order in seasonal differencing, **Q** is the number of order in the moving average seasonality and **S** is the number of period in the season.

We are already familiar with how to determine the (p,d,q), it exactly same as what with did in ARIMA. 

To determine the (P,D,Q) the process is also same By looking at ACF we see the spike at lag=1 hense we can coclude the AR seasonality is 1, we previusly performed the differencing and already know we need one seasonal differncing. 
By looking at PACF we noticed the first spike at lag=1, so we can conclude that the number of MA seasonality is also 1.
and the number of periods in season is basiccaly 365, as we have daily dataset.
I have used 365 as the number of perid at first but the fitting model was too slow, hence I resampled the data to monthly and as aresult determined the number of period as 12.

````python
from sklearn.metrics import mean_squared_error


def validate(data, perc):
    predictions = []

    train, test = train_test_split(data, perc)

    history = [x for x in train]

    for i in range(len(test)):
        test_X, test_y = test[i, :-1], test[i, -1]

        pred = xgb_predict(history, test_X[0])
        predictions.append(pred)

        history.append(test[i])

    error = mean_squared_error(test[:, -1], predictions, squared=False)

    return error, test[:, -1], predictions
````

## TimeseriesGenerator

Before fitting a supervised learning model, we have to transform the time series into a structure of samples with input and output components.

Implementing this transformation manually could be challenging. hence we have used the TimeseriesGenerator in the Keras deep learning library which automatically transforms both univariate and multivariate time series data into samples that can be used to train deep learning models.
First, we define the TimeseriesGenerator and then we use it to train the models.

````python
from keras.preprocessing.sequence import TimeseriesGenerator
````

Here we **Defining the TimeseriesGenerator**,
we built an instance of the class and define the input and output elements of the time series problem, as we can see it returns an instance of the Sequence class, which we can use to iterate over the series' inputs and outputs.

The input and output series in most time series prediction problems are the same, and in our case are also the same. 

The class isn't technically a generator because it isn't a Python Generator and we can't use the next() function on it.

There are certainly additional factors that we need to configure in addition to determining the input and output components of the time series problem; some of the factors that we used here length and batch size.

**length** is basically the number of lag observations that we decide to use in the input of every sample, here we defined a variable `n_input = 30` and assigned it to the lenght parameter. I sat it to 30 beacuse we want to predict 30 days ahead, and here we are predicting with one step(n_features=1), so my logic is 30 days ahead simply means we need to take 29 lags pbservations to predict the 30rd one.

**Batch size** is simply the number of training samples in one batch. not that the terms batch size and number of the batch are not the same.
The reason behind this term is that we can't feed the whole dataset into the neural network at once, hence we split it into specific numbers of batches. It could be any number depending on the dataset, in this study I first chose batche_size=1 , 2, 8 and 16 but they were too slow and I changed to 32 which is one of the popular batch sizes and it speeds up the modeling.

# Error metrics



Let us first review the error metrics in Python before going into the concept of RMSE.

Error metrics allow us to monitor efficiency and accuracy using a variety of metrics, such as **Mean Square Error(MSE), Root Mean Square Error(RMSE),R-square, Accuracy, MAPE, etc.

**Mean Square Error(MSE) ** is used for evaluating the accuracy and error rate of any machine learning algorithm for a regression problem. hence we can say MSE is a risk function that allows us to calculate the average squared difference between a feature's predicted and actual value.

**Root Mean Square Error (RMSE)** is the square root of the value received by the Mean Square Error function.

We may easily depict the difference between estimated and real values of a model parameter using RMSE. This allows us to assess the model's efficiency.

A RMSE score of less than 180 is usually assumed good for a moderately or well-functioning algorithm. For the scores greater than 180, we must undertake feature selection and hyperparameter tuning on the model's parameters.
